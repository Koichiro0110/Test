```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ax = plt.figure().add_subplot(111, projection = '3d')

E_0 = 0.
W = -0.12
a = 1.
a1 = a
a2 = a

x = np.linspace(-4*np.pi/(3*a), 4*np.pi/(3*a), 100)
y = np.linspace(-2*np.pi/np.sqrt(3)*a, 2*np.pi/np.sqrt(3)*a, 100)
x, y = np.meshgrid(x, y)

def k1a1(x,y):
    dot = a*(1/2*x + np.sqrt(3)/2*y)
    return dot

def k2a2(x,y):
    dot = a*(-1/2*x + np.sqrt(3)/2*y)
    return dot
    
def E1(x,y):
    E = E_0 + W*np.sqrt(3 + 2*(np.cos(k1a1(x,y)) + np.cos(k2a2(x,y)) + np.cos(k1a1(x,y) - k2a2(x,y))) )
    return E

def E2(x,y):
    E = E_0 - W*np.sqrt(3 + 2*(np.cos(k1a1(x,y)) + np.cos(k2a2(x,y)) + np.cos(k1a1(x,y) - k2a2(x,y))) )
    return E

ax.plot_surface(y, x, E1(x,y), color = 'royalblue', edgecolor ='royalblue', 
                     lw = 0.5, rstride = 8, cstride = 8, alpha = 0.3)
ax.plot_surface(y, x, E2(x,y), color = 'mediumspringgreen', edgecolor ='mediumspringgreen', 
                     lw = 0.5, rstride = 8, cstride = 8, alpha = 0.3)

ax.set_xlabel('ky')
ax.set_ylabel('kx')
ax.set_zlabel('E(kx,ky)')
ax.set_title("Graphene's band structure in reciprocal space")

from matplotlib.backends.backend_pdf import PdfPages
pdf = PdfPages('Graphen Band.pdf')
pdf.savefig()
pdf.close()
```

```python
ax = plt.figure().add_subplot(111, projection = '3d')

E_2 = 0.
Ws = -0.12
Wpz = -0.01
a = 1.
a1 = a
a2 = a

x = np.linspace(-4*np.pi/(3*a), 4*np.pi/(3*a), 50)
y = np.linspace(-2*np.pi/np.sqrt(3)*a, 2*np.pi/np.sqrt(3)*a, 50)
x, y = np.meshgrid(x, y)

def k1a1(x,y):
    dot = a*(1/2*x + np.sqrt(3)/2*y)
    return dot

def k2a2(x,y):
    dot = a*(-1/2*x + np.sqrt(3)/2*y)
    return dot
    
def E1(x,y,E0,W):
    E = E0 + W*np.sqrt(3 + 2*(np.cos(k1a1(x,y)) + np.cos(k2a2(x,y)) + np.cos(k1a1(x,y) - k2a2(x,y))) )
    return E

def E2(x,y,E0,W):
    E = E0 - W*np.sqrt(3 + 2*(np.cos(k1a1(x,y)) + np.cos(k2a2(x,y)) + np.cos(k1a1(x,y) - k2a2(x,y))) )
    return E

Es1 = E1(x,y,E_2,Ws)
Es2 = E2(x,y,E_2,Ws)
Epz1 = E1(x,y,E_2,Wpz)
Epz2 = E2(x,y,E_2,Wpz)



ax.plot_surface(y, x, Es1, color = 'royalblue', edgecolor ='royalblue', 
                     lw = 0.5, rstride = 8, cstride = 8, alpha = 0.3)
ax.plot_surface(y, x, Es2, color = 'mediumspringgreen', edgecolor ='mediumspringgreen', 
                     lw = 0.5, rstride = 8, cstride = 8, alpha = 0.3)
ax.plot_surface(y, x,Epz1, color = 'red', edgecolor ='red', 
                     lw = 0.5, rstride = 8, cstride = 8, alpha = 0.3)
ax.plot_surface(y, x, Epz2, color = 'orange', edgecolor ='orange', 
                     lw = 0.5, rstride = 8, cstride = 8, alpha = 0.3)

ax.set_xlabel('ky')
ax.set_ylabel('kx')
ax.set_zlabel('E(kx,ky)')
ax.set_title("Graphene's s-band and pz-band in reciprocal space")

from matplotlib.backends.backend_pdf import PdfPages
pdf = PdfPages('Graphen s,pz-band.pdf')
pdf.savefig()
pdf.close()
```
