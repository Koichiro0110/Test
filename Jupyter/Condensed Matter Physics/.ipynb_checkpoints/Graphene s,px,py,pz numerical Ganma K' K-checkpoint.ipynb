```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.integrate import tplquad
```

```python
# physical parameters

# Bohr radius
a0 = 0.529 # Å

# Vaccume permitivity
e_0 = 8.854e-42 # kg^-1 * Å^-3 * s^4 * A^2

# Elementary charge
e = 1.60217663e-19 # A * s 

# Carbon's mass ratio
Z = 12.01

# Lattice Spacing of the Graphene
d = 1.42 # Å

# Lattice constant of the Bravais
a = np.sqrt(3)*d

# 1[eV] = 1.6*10^(-19) [J]
# 1[J] = 1/1.6*10^(19)[eV]
# 1[J] = kg * m^2 * s^(-2) = 10^20 * kg * s^(-2) * Å^2

# So 1 [kg * s^(-2) * Å^2] = 1/16 [eV]
```

```python
# Define the dot product

def k1a1(x,y):
    dot = a*(1/2*x + np.sqrt(3)/2*y)
    return dot

def k2a2(x,y):
    dot = a*(-1/2*x + np.sqrt(3)/2*y)
    return dot

# Define the matrix (Hamiltonian)

def matrix(x,y,eps0s,eps0p,sssigma,spsigma,ppsigma,pppi):
    
    I = np.zeros((6,6))
    I[0][0] = - eps0s
    I[0][1] = -sssigma*(1 + np.exp(1j*(k1a1(x,y))) + np.exp(1j*(k2a2(x,y))))
    I[0][2] = 0
    I[0][3] = -np.sqrt(3)/2*spsigma*(np.exp(1j*(k1a1(x,y))) + np.exp(1j*(k2a2(x,y))))
    I[0][4] = 0
    I[0][5] = spsigma - 1/2*spsigma*(np.exp(1j*(k1a1(x,y))) + np.exp(1j*(k2a2(x,y))))

    I[1][0] = -sssigma*(1 + np.exp(-1j*(k1a1(x,y))) + np.exp(-1j*(k2a2(x,y))))
    I[1][1] = - eps0s
    I[1][2] = np.sqrt(3)/2*spsigma*(np.exp(-1j*(k1a1(x,y))) + np.exp(-1j*(k2a2(x,y))))
    I[1][3] = 0
    I[1][4] = -spsigma + 1/2*spsigma*(np.exp(-1j*(k1a1(x,y))) + np.exp(-1j*(k2a2(x,y))))
    I[1][5] = 0

    I[2][0] = 0
    I[2][1] = np.sqrt(3)/2*spsigma*(np.exp(1j*(k1a1(x,y))) + np.exp(1j*(k2a2(x,y))))
    I[2][2] = - eps0p
    I[2][3] = -pppi - 1/4*(3*ppsigma + 1*pppi)*(np.exp(1j*(k1a1(x,y))) + np.exp(1j*(k2a2(x,y))))
    I[2][4] = 0
    I[2][5] = -np.sqrt(3)/4*(ppsigma-pppi)*(np.exp(1j*(k1a1(x,y))) - np.exp(1j*(k2a2(x,y))))

    I[3][0] = -np.sqrt(3)/2*spsigma*(np.exp(-1j*(k1a1(x,y))) + np.exp(-1j*(k2a2(x,y))))
    I[3][1] = 0
    I[3][2] = -pppi - 1/4*(3*ppsigma + 1*pppi)*(np.exp(-1j*(k1a1(x,y))) + np.exp(-1j*(k2a2(x,y))))
    I[3][3] = - eps0p
    I[3][4] = -np.sqrt(3)/4*(ppsigma-pppi)*(np.exp(-1j*(k1a1(x,y))) + np.exp(-1j*(k2a2(x,y))))
    I[3][5] = 0

    I[4][0] = 0
    I[4][1] = -spsigma + 1/2*spsigma*(np.exp(1j*(k1a1(x,y))) + np.exp(1j*(k2a2(x,y))))
    I[4][2] = 0
    I[4][3] = -np.sqrt(3)/4*(ppsigma-pppi)*(np.exp(1j*(k1a1(x,y))) + np.exp(1j*(k2a2(x,y))))
    I[4][4] = - eps0p
    I[4][5] = -ppsigma -1/4*(ppsigma+pppi)*(np.exp(1j*(k1a1(x,y))) + np.exp(1j*(k2a2(x,y))))

    I[5][0] = spsigma - 1/2*spsigma*(np.exp(-1j*(k1a1(x,y))) + np.exp(-1j*(k2a2(x,y))))
    I[5][1] = 0
    I[5][2] = -np.sqrt(3)/4*(ppsigma-pppi)*(np.exp(-1j*(k1a1(x,y))) - np.exp(-1j*(k2a2(x,y))))
    I[5][3] = 0
    I[5][4] = -ppsigma -1/4*(ppsigma+pppi)*(np.exp(-1j*(k1a1(x,y))) + np.exp(-1j*(k2a2(x,y))))
    I[5][5] = - eps0p
    
    return I

# Define the matrix (Hamiltonian)

def Matrix(x,y,eps0p,pppi):
    
    I = np.zeros((2,2))
    I[0][0] = - eps0p
    I[0][1] = -pppi*(1 + np.exp(1j*(k1a1(x,y))) + np.exp(1j*(k2a2(x,y))))
    I[1][0] = -pppi*(1 + np.exp(-1j*(k1a1(x,y))) + np.exp(-1j*(k2a2(x,y))))
    I[1][1] = - eps0p

    
    return I
```

```python
# s-orbital
def s_orbit(x, y, z, a0 = a0, Z = Z):
    r = np.sqrt(x**2 + y**2 + z**2 + 1e-20)
    psi = 1/np.sqrt(np.pi) * (Z/(2*a0))**(3/2) * (1 - Z*r/(2*a0)) * np.exp(-Z*r/(2*a0))
    return psi

# pz-orbital
def pz_orbit(x, y, z, a0 = a0, Z = Z):
    r = np.sqrt(x**2 + y**2 + z**2 + 1e-20)
    cos_theta = z/(r+1e-20)
    psi = 1/(2*np.sqrt(np.pi)) * (Z/(2*a0))**(3/2) * (Z*r/a0) * np.exp(-Z*r/(2*a0)) * cos_theta
    return psi
    
# px-orbital
def px_orbit(x, y, z, a0 = a0, Z = Z):
    r = np.sqrt(x**2 + y**2 + z**2 + 1e-20)
    R = np.sqrt(x**2 + y**2 + 1e-40)
    sin_theta = R/(r+1e-20)
    cos_phi = x/(r*sin_theta)
    psi = - 1/(2*np.sqrt(np.pi)) * (Z/(2*a0))**(3/2) * (Z*r/a0) * np.exp(-Z*r/(2*a0)) * sin_theta * cos_phi
    return psi

# py-orbital
def py_orbit(x, y, z, a0 = a0, Z = Z):
    r = np.sqrt(x**2 + y**2 + z**2 + 1e-20)
    R = np.sqrt(x**2 + y**2 + 1e-40)
    sin_theta = R/(r+1e-20)
    sin_phi = y/(r*sin_theta)
    psi = - 1/(2*np.sqrt(np.pi)) * (Z/(2*a0))**(3/2) * (Z*r/a0) * np.exp(-Z*r/(2*a0)) * sin_theta * sin_phi
    return psi
```

```python
# Coulomb potential
def deltaU(x, y, z, Z = Z): # Spherically symmetric
    r = np.sqrt(x**2 + y**2 + z**2 + 1e-20)
    potential = - 1/(4*np.pi*e_0)*Z*e**2/np.abs(r)
    return potential
```

```python
def deltaU_polar(r, Z = Z): # Spherically symmetric
    potential = - 1/(4*np.pi*e_0)*Z*e**2/np.abs(r+1e-20)
    return potential
```

```python
r = np.linspace(1e-20,a0*1e-18,100000) # This is in Armstrong!!!
# So Almost a delta fucntion in Armstrong order.

fig, ax = plt.subplots(1, 2, figsize = (12, 6))
ax[0].plot(r, deltaU_polar(r, Z = Z))

r = np.linspace(1e-20,a0,100000)
ax[1].plot(r, deltaU_polar(r, Z = Z))
```

```python
x = np.linspace(1e-20,a0*4+d,100000) # This is in Armstrong!!!
y = np.linspace(1e-20,a0*4+d,100000)
z = np.linspace(1e-20,a0*4+d,100000)


fig, ax = plt.subplots(1, 4, figsize = (12, 6))
ax[0].set_xlabel("[Å]")
ax[1].set_xlabel("[Å]")
ax[2].set_xlabel("[Å]")
ax[0].plot(x, s_orbit(x, 0, 0), label = "s-orbital")
ax[1].plot(x, s_orbit(x, 0, 0)* deltaU(x - d, 0, 0)* s_orbit(x - d, 0, 0), label = "two-center integral function")
ax[2].plot(x, deltaU(x - d, 0, 0))


ax[0].plot([0, 0], [s_orbit(0,0,0), s_orbit(0,0,0)], color = "lime")

#ax[1].plot(r, deltaU_polar(r, Z = Z))
fig.tight_layout()    
```

```python
# Calculate the hopping integrations.
L = 4*a0 + d
func = lambda x,y,z:s_orbit(x, y, z)* deltaU(x - d, y, z)* s_orbit(x - d, y, z)
sssigmaintegral = tplquad(func, -L, L, -L, L, -L, L)
sssigma = sssigmaintegral[0]/16
print("sssigma = ", sssigma, "[eV]")

L = 4*a0 + d
func = lambda x,y,z:s_orbit(x, y, z)* deltaU(x - d, y, z)* px_orbit(x - d, y, z)
spsigmaintegral = tplquad(func, -L, L, -L, L, -L, L)
spsigma = spsigmaintegral[0]/16
print("spsigma = ", spsigma, "[eV]")

L = 4*a0 + d
func = lambda x,y,z:px_orbit(x, y, z)* deltaU(x - d, y, z)* px_orbit(x - d, y, z)
ppsigmaintegral = tplquad(func, -L, L, -L, L, -L, L)
ppsigma = ppsigmaintegral[0]/16
print("ppsigma = ", ppsigma, "[eV]")

L = 4*a0 + d
func = lambda x,y,z:pz_orbit(x, y, z)* deltaU(x - d, y, z)* pz_orbit(x - d, y, z)
pppiintegral = tplquad(func, -L, L, -L, L, -L, L)
pppi = pppiintegral[0]/16
print("pppi = " ,pppi, "[eV]")
```

```python
# Set the energy of atomic orbital as 0 [eV].
eps0s = 0 #[eV]
eps0p = 0 #[eV]
```

```python
# Solve the eigenvalue equation along a certain path in Brillouin Zone.
N = 1000
x_values1 = np.linspace(0, 4*np.pi/(3*a), 2*N)
y_values1 = np.zeros(2*N)
x_values1plot = x_values1

x2 = np.linspace(2*np.pi/(3*a), 4*np.pi/(3*a), N)
x_values2 = x2[::-1]
x_values2plota = np.linspace(0, 2*np.pi/(3*a), N) + 4*np.pi/(3*a)*np.ones(N)
x_values2plot = x_values2plota[1::]
y_values2 = -np.sqrt(3)*x_values2 + 4*np.pi/(np.sqrt(3)*a)

x3 = np.linspace(0, 2*np.pi/(3*a), N)
x_values3 = x3[::-1]
x_values3plota = np.linspace(0, 2*np.pi/(3*a), N) + 6*np.pi/(3*a)*np.ones(N)
x_values3plot = x_values3plota[1::]

y_values3 = np.sqrt(3)*x_values3

x_valuesplot = np.concatenate([x_values1plot, x_values2plot, x_values3plot])

band = []
for i,j in zip(x_values1,y_values1):
    eigen1 = np.linalg.eigh(matrix(i,j,eps0s,eps0p,sssigma,spsigma,ppsigma,pppi))
    eigen2 = np.linalg.eigh(Matrix(i,j,eps0p,pppi))
    eigen = np.concatenate([eigen1[0], eigen2[0]])
    band.append(eigen)
    
for i,j in zip(x_values2[1::],y_values2[1::]):
    eigen1 = np.linalg.eigh(matrix(i,j,eps0s,eps0p,sssigma,spsigma,ppsigma,pppi))
    eigen2 = np.linalg.eigh(Matrix(i,j,eps0p,pppi))
    eigen = np.concatenate([eigen1[0], eigen2[0]])
    band.append(eigen)
    
for i,j in zip(x_values3[1::],y_values3[1::]):
    eigen1 = np.linalg.eigh(matrix(i,j,eps0s,eps0p,sssigma,spsigma,ppsigma,pppi))
    eigen2 = np.linalg.eigh(Matrix(i,j,eps0p,pppi))
    eigen = np.concatenate([eigen1[0], eigen2[0]])
    band.append(eigen)

# Transpose to plot
bandT = np.transpose(band)
```

```python
#Brillouin Zoneの軌跡をプロットする！！
figure = plt.figure(figsize = (9,7))
axes = figure.gca()

x_values = np.concatenate([x_values1, x_values2[1::], x_values3[1::]])
y_values = np.concatenate([y_values1, y_values2[1::], y_values3[1::]])

axes.scatter(x_values, y_values, s = 1)
axes.scatter(x_values1[0],y_values1[0] , s = 100, label = " $ \Gamma $ ")
axes.scatter(x_values2[0],y_values2[0] , s = 100, label = " $ K' $ ")
axes.scatter(x_values3[0],y_values3[0] , s = 100, label = " $ K $ ")

axes.set_title("Reciprocal Space", fontsize = 22)

axes.set_xlabel("$k_{x}~[~Å^{-1}~]$", fontsize = 16)
axes.set_xticks([0, 2*np.pi/(3*a), 4*np.pi/(3*a)])
ticks = ["$0$", r"$\frac{2 \pi}{3a}$", r"$\frac{4 \pi}{3a}$"]
axes.set_xticklabels(ticks,fontsize = 20)

axes.set_ylabel("$k_{y}~[~Å^{-1}~]$", fontsize = 16)
axes.set_yticks([0, np.pi/(np.sqrt(3)*a), 2*np.pi/(np.sqrt(3)*a)])
ticks = [r"$0$", r"$\frac{\pi}{ \sqrt{3} a}$", r"$\frac{2 \pi}{ \sqrt{3} a}$"]
axes.set_yticklabels(ticks,fontsize = 20)

axes.legend(fontsize = 18)
```

```python
# Make a plot of the band structure.
fig1 = plt.figure(figsize = (9,7))
ax = fig1.gca()

for i in range(8):
    if i == 6:
        ax.plot(x_valuesplot, bandT[i], label = "pz-band", color = "red") # Converting into eV but not sure if it's right)
    elif i == 7:
        ax.plot(x_valuesplot, bandT[i], color = "red") # Converting into eV but not sure if it's right)
    else:
        ax.plot(x_valuesplot, bandT[i], color = "black") # Converting into eV but not sure if it's right)

ax.plot([4*np.pi/(3*a), 4*np.pi/(3*a)], [-bandT[0][0], bandT[0][0]], color = "lime", label = "Dirac points", linewidth = 1)
ax.plot([6*np.pi/(3*a), 6*np.pi/(3*a)], [-bandT[0][0], bandT[0][0]], color = "lime", linewidth = 1)

ax.set_title("Energy band structure of Graphene s,px,py,pz", fontsize = 22)
ax.set_xlabel("Reciprocal space [ $ Å^{-1} $ ]", fontsize = 16)
ax.set_ylabel("Energy band [eV]", fontsize = 16)

ticks = ["$ \Gamma $", "$ K' $", "$ K $", "$ \Gamma $"]
ax.set_xticks([0, 4*np.pi/(3*a), 6*np.pi/(3*a), 8*np.pi/(3*a)])
ax.set_xticklabels(ticks,fontsize = 16)
#ax.set_yticks([0])
#ax.set_yticklabels(["$0$"],fontsize = 16)
ax.legend(fontsize = 13)

fig2 = plt.figure(figsize = (9,7))
ax = fig2.gca()

indexes = [6,7]

for i in indexes:
    ax.plot(x_valuesplot, bandT[i], color = "red") # Converting into eV but not sure if it's right)
    
ax.plot([4*np.pi/(3*a), 4*np.pi/(3*a)], [-bandT[6][0], bandT[6][0]], color = "lime", label = "Dirac points", linewidth = 1)
ax.plot([6*np.pi/(3*a), 6*np.pi/(3*a)], [-bandT[6][0], bandT[6][0]], color = "lime", linewidth = 1)

ax.set_title("Graphene's pz-band structure", fontsize = 22)
ax.set_xlabel("Reciprocal space [ $ Å^{-1} $ ]", fontsize = 16)
ax.set_ylabel("Energy band [eV]", fontsize = 16)

ticks = ["$ \Gamma $", "$ K' $", "$ K $", "$ \Gamma $"]
ax.set_xticks([0, 4*np.pi/(3*a), 6*np.pi/(3*a), 8*np.pi/(3*a)])
ax.set_xticklabels(ticks,fontsize = 16)
#ax.set_yticks([0])
#ax.set_yticklabels(["$0$"],fontsize = 16)
ax.legend(fontsize = 13)

from matplotlib.backends.backend_pdf import PdfPages
pdf = PdfPages('Energy bands of Graphene s,px,py,pz.pdf')
pdf.savefig(fig1)
pdf.savefig(fig2)
pdf.close()
```
