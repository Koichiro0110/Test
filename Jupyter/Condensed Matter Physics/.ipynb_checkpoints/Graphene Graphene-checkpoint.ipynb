```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
from scipy.interpolate import interp1d
smp.init_printing()
```

```python
epsilon, eps0s, eps0p = smp.symbols('epsilon epsilon_0s epsilon_0p')
k1,k2 = smp.symbols('k_1,k_2,')
a1,a2 = smp.symbols('a_1, a_2')
sssigma, spsigma, ppsigma, pppi = smp.symbols('ss_sigma sp_sigma pp_sigma pp_pi')
```

```python
# Genral matrix
def general_matrix(m, n, s):
    smp.Symbol(s)
    elements = lambda i, j : smp.var('{}{}{}'.format(s, i+1, j+1))
    return smp.Matrix(m, n, elements)

Matrix = general_matrix(8,8,"M")
display(Matrix)
```

```python
I = smp.eye(6)
I
```

```python
I[0,0] = epsilon - eps0s
I[0,1] = -sssigma*(1 + smp.exp(smp.I*(k1*a1)) + smp.exp(smp.I*(k2*a2)))
I[0,2] = 0
I[0,3] = -smp.sqrt(3)/2*spsigma*(smp.exp(smp.I*(k1*a1)) + smp.exp(smp.I*(k2*a2)))
I[0,4] = 0
I[0,5] = spsigma - 1/2*spsigma*(smp.exp(smp.I*(k1*a1)) + smp.exp(smp.I*(k2*a2)))
```

```python
I[1,0] = -sssigma*(1 + smp.exp(-smp.I*(k1*a1)) + smp.exp(-smp.I*(k2*a2)))
I[1,1] = epsilon - eps0s
I[1,2] = smp.sqrt(3)/2*spsigma*(smp.exp(-smp.I*(k1*a1)) + smp.exp(-smp.I*(k2*a2)))
I[1,3] = 0
I[1,4] = -spsigma + 1/2*spsigma*(smp.exp(-smp.I*(k1*a1)) + smp.exp(-smp.I*(k2*a2)))
I[1,5] = 0
```

```python
I[2,0] = 0
I[2,1] = smp.sqrt(3)/2*spsigma*(smp.exp(smp.I*(k1*a1)) + smp.exp(smp.I*(k2*a2)))
I[2,2] = epsilon - eps0p
I[2,3] = -pppi - 1/4*(3*ppsigma + 1*pppi)*(smp.exp(smp.I*(k1*a1)) + smp.exp(smp.I*(k2*a2)))
I[2,4] = 0
I[2,5] = -smp.sqrt(3)/4*(ppsigma-pppi)*(smp.exp(smp.I*(k1*a1)) - smp.exp(smp.I*(k2*a2)))
```

```python
I[3,0] = -smp.sqrt(3)/2*spsigma*(smp.exp(-smp.I*(k1*a1)) + smp.exp(-smp.I*(k2*a2)))
I[3,1] = 0
I[3,2] = -pppi - 1/4*(3*ppsigma + 1*pppi)*(smp.exp(-smp.I*(k1*a1)) + smp.exp(-smp.I*(k2*a2)))
I[3,3] = epsilon - eps0p
I[3,4] = -smp.sqrt(3)/4*(ppsigma-pppi)*(smp.exp(-smp.I*(k1*a1)) + smp.exp(-smp.I*(k2*a2)))
I[3,5] = 0
```

```python
I[4,0] = 0
I[4,1] = -spsigma + 1/2*spsigma*(smp.exp(smp.I*(k1*a1)) + smp.exp(smp.I*(k2*a2)))
I[4,2] = 0
I[4,3] = -smp.sqrt(3)/4*(ppsigma-pppi)*(smp.exp(smp.I*(k1*a1)) + smp.exp(smp.I*(k2*a2)))
I[4,4] = epsilon - eps0p
I[4,5] = -ppsigma -1/4*(ppsigma+pppi)*(smp.exp(smp.I*(k1*a1)) + smp.exp(smp.I*(k2*a2)))
```

```python
I[5,0] = spsigma - 1/2*spsigma*(smp.exp(-smp.I*(k1*a1)) + smp.exp(-smp.I*(k2*a2)))
I[5,1] = 0
I[5,2] = -smp.sqrt(3)/4*(ppsigma-pppi)*(smp.exp(-smp.I*(k1*a1)) - smp.exp(-smp.I*(k2*a2)))
I[5,3] = 0
I[5,4] = -ppsigma -1/4*(ppsigma+pppi)*(smp.exp(-smp.I*(k1*a1)) + smp.exp(-smp.I*(k2*a2)))
I[5,5] = epsilon - eps0p
```
