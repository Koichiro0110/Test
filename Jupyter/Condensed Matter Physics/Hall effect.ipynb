```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as spi
import matplotlib.patches as pt
```

```python
# ODEs

def electron(y, t, param):
    d_var = np.zeros(4)
    e, E, B, m = param
    
    d_var[0] = y[2]
    d_var[1] = y[3]
        
    d_var[2] = -e*E/m + -e*B/m * (y[3])
    d_var[3] = e*B/m * (y[2])
            
    return d_var

# Value of parameters
e = 1
E = 1
B = 1
m = 1
parameters = [e, E, B, m]

y_init = np.zeros(4)

t_end = 30
N = 128
t = np.linspace(0, t_end, N)

# Solve

solution = spi.odeint(electron, y_init, t, args = (parameters,))
solutionT = np.transpose(solution)

# Plot

fig1 = plt.figure(figsize = (12, 9))
plt.title("Electron")
plt.xlabel("time (sec)")
plt.ylabel("position")
plt.plot(t, solutionT[0], label = "$x$")
plt.plot(t, solutionT[1], label = "$y$")
plt.legend()
plt.show()
```
