```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
E = np.linspace(0,10, 1000)

def D1(E):
    m = 1
    hbar = 1
    d = 1
    C = 2
    D = C*d/2* (np.sqrt(2*m/(hbar**2)))**(d) * np.sqrt(E)**(d-2)
    return D

def D2(E):
    m = 1
    hbar = 1
    d = 2
    C = np.pi
    D = C*d/2* (np.sqrt(2*m/(hbar**2)))**(d) * np.sqrt(E)**(d-2)
    return D
    
def D3(E):
    m = 1
    hbar = 1
    d = 3
    C = 4*np.pi/3
    D = C*d/2* (np.sqrt(2*m/(hbar**2)))**(d) * np.sqrt(E)**(d-2)
    return D
  
fig = plt.figure(figsize = (9,7))
ax = fig.gca()

ax.plot(E, D1(E), label = "d = 1")
ax.plot(E, D2(E), label = "d = 2")
ax.plot(E, D3(E), label = "d = 3")
    
ax.set_title("DOS of Free Electron", fontsize = 22)
ax.set_xlabel("Energy, E", fontsize = 16)
ax.set_ylabel("Density of Energy, D(E)", fontsize = 16)
ax.legend(fontsize = 18)

from matplotlib.backends.backend_pdf import PdfPages
pdf = PdfPages('DOS of Free Electron.pdf')
pdf.savefig()
pdf.close()
```
