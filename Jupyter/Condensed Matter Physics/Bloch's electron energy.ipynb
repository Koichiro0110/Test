# Bloch's electron energy
Bloch's electron energy

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 1
m = 1
hbar = 1

def k(E, lamda):
    k = 1/a* np.arccos( np.cos(np.sqrt(2*m*E)*a/hbar) + lamda*np.sqrt(m)/(hbar*np.sqrt(2*E))*np.sin(np.sqrt(2*m*E)*a/hbar) )
    return k
```

```python
fig = plt.figure(figsize = (10,8))
ax = fig.gca()

E = np.linspace(-2*np.pi/a, 2*np.pi/a, 1000)
values = [-100, -10, -1, 1, 10, 100]
for lamda in values:
    ax.plot(E, k(E, lamda), label = f"lamda = {lamda}")
ax.set_xlabel("Energy", fontsize = 20)
ax.set_ylabel("k", fontsize = 20)
ax.legend()
```
