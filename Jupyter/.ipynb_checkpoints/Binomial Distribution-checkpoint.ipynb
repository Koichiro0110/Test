```python
import numpy as np
from scipy.special import factorial
import matplotlib.pyplot as plt
```

```python
# Define Binomilal Distribution
def Binomial_Distribution(n, N, p):
    PN = factorial(N)/(factorial(N - n)*factorial(n))*(p**n)*((1-p)**(N-n))
    return PN
```

```python
# Create a set of N values.
Nvalues = [3, 4, 5, 10, 20]

# Get a figure where I plot.
fig = plt.figure(figsize = (10,8))

# Get the axes.
axes = fig.gca()

# Set the axes and the title as you like.
axes.set_xticks([i for i in range(Nvalues[-1]+1)])

axes.set_title("Binomial Distribution", fontsize = 22)
axes.set_xlabel("n", fontsize = 16)
axes.set_ylabel(r"$P_{N}(n)$", fontsize = 16)

# Make plots.
for N in Nvalues:
    n = np.linspace(0., N, 1000)
    PN = Binomial_Distribution(n, N, 1/2)
    axes.plot(n, PN, label = f"N = {N}")
    
# Show the legend.
axes.legend(fontsize = 18)
```

```python
Nvalues = [3, 4, 5, 10, 20]

fig = plt.figure(figsize = (10,8))
axes = fig.gca()
axes.set_xticks([i for i in range(Nvalues[-1]+1)])
axes.set_title("Binomial Distribution", fontsize = 22)
axes.set_xlabel("n", fontsize = 16)
axes.set_ylabel(r"$P_{N}(n)$", fontsize = 16)

for N in Nvalues:
    n = np.linspace(0., N, N) # Only plot at integer n.
    PN = Binomial_Distribution(n, N, 1/2)
    axes.plot(n, PN, linestyle = "None", marker = "o", label = f"N = {N}") #linestyle = "None" makes "dot" plots.
axes.legend(fontsize = 18)
```

```python
# Define Binomilal Distribution for large N
def Binomial_Distribution_LargeN(n, N, p):
    x = n/N
    PN = np.sqrt(1/(2*np.pi*N*x*(1-x)))*np.exp(-N*( x*np.log(x/p) + (1-x)*np.log((1-x)/(1-p)) ) )
    return PN
```

```python
# Create a set of large N values.
Nvalues = [1e4, 3e4, 5e4, 10e4, 20e4]

# Get a figure where I plot.
fig = plt.figure(figsize = (10,8))

# Get the axes.
axes = fig.gca()

axes.set_title("Binomial Distribution", fontsize = 22)
axes.set_xlabel("n", fontsize = 16)
axes.set_ylabel(r"$P_{N}(n)$", fontsize = 16)

# Make plots.
for N in Nvalues:
    n = np.linspace(0.01, N-1, 1000)
    PN = Binomial_Distribution_LargeN(n, N, 1/2)
    axes.plot(n, PN, label = f"N = {N}")
    
# Show the legend.
axes.legend(fontsize = 18)
```
