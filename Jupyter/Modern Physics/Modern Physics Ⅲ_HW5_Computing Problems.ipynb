```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline
```

```python
#Reading all the csv files
hydrogen1 = pd.read_csv("hydrogen_1.csv")#un-saturated
hydrogen2 = pd.read_csv("hydrogen_2.csv")
deuterium1 = pd.read_csv("deuterium_1.csv")#un-saturated
deuterium2 = pd.read_csv("deuterium_2.csv")
neon1 = pd.read_csv("neon_1.csv")#un-saturated
```

```python
#Changing all their columns
hydrogen1.columns = ["WV", "IN"]
hydrogen1.columns = ["WV", "IN"]
hydrogen2.columns = ["WV", "IN"]
deuterium1.columns = ["WV", "IN"]
deuterium2.columns = ["WV", "IN"]
neon1.columns = ["WV", "IN"]
```

```python
# [2]

#Labeling the data
d1x = deuterium1["WV"]
d1y = deuterium1["IN"]
h1x = hydrogen1["WV"]
h1y = hydrogen1["IN"]

#Plotting the unsaturated spectra of hydrogen and deuterium
fig = plt.figure(figsize = (10,8))
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(d1x, d1y, label = "deutrium", color = "lime")
axes.plot(h1x, h1y, label = "hydrogen", color = "blue")

#Labeling other properties
axes.set_title("Curves of the 'un-saturated' spectra", fontsize = 24)
axes.set_xlabel("Wavelength[$nm$]", fontsize = 20)
axes.set_ylabel("Intensity[$W/m^2$]", fontsize = 20)
axes.legend(fontsize = 18)

```

```python
# [4]

#Labeling the data
d1x = deuterium1["WV"]
d1y = deuterium1["IN"]
h1x = hydrogen1["WV"]
h1y = hydrogen1["IN"]

#Plotting the unsaturated spectra of hydrogen and deuterium
fig = plt.figure(figsize = (10,8))
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(d1x, d1y, label = "deutrium", color = "lime")
axes.plot(h1x, h1y, label = "hydrogen", color = "blue")

#Overlay (visible) Balmer series in the spectrum

Balmer = [656, 486, 434, 410]

for i in range(4):
    axes.plot([Balmer[i], Balmer[i]], [0, 65000], label = r"Balmer n="+str(i+3)+"2")

#Finding peaks (including noises) and plotting markers
maxid_d1y = signal.argrelmax(d1y.to_numpy(), order = 45)
maxid_h1y = signal.argrelmax(h1y.to_numpy(), order = 45)

axes.plot(d1x.to_numpy()[maxid_d1y], d1y.to_numpy()[maxid_d1y], "r+", markersize = "15", label = "deutrium peak")
axes.plot(h1x.to_numpy()[maxid_h1y], h1y.to_numpy()[maxid_h1y], "bo", markersize = "5", label = "hydrogen peak")

#Labeling other properties
axes.set_title("Curves of the 'un-saturated' spectra", fontsize = 24)
axes.set_xlabel("Wavelength[$nm$]", fontsize = 20)
axes.set_ylabel("Intensity[(Arbitrary)$W/m^2$]", fontsize = 20)
axes.legend(fontsize = 18)

Peaks = d1x.to_numpy()[maxid_d1y]
Intensity_Peaks = d1y.to_numpy()[maxid_d1y]
print(Peaks[6], Intensity_Peaks[6])
print(Peaks[7], Intensity_Peaks[7])
print(Peaks[9], Intensity_Peaks[9])
print(Peaks[14], Intensity_Peaks[14])

Peaks = h1x.to_numpy()[maxid_h1y]
Intensity_Peaks = h1y.to_numpy()[maxid_h1y]
print(Peaks[6], Intensity_Peaks[6])
print(Peaks[7], Intensity_Peaks[7])
print(Peaks[9], Intensity_Peaks[9])
print(Peaks[13], Intensity_Peaks[13])
```

```python
#[5]

#Labeling the data
n1x = neon1["WV"]
n1y = neon1["IN"]


#Plotting the unsaturated spectra of neon
fig = plt.figure(figsize = (10,8))
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(n1x, n1y, label = "neon", color = "skyblue")

#Wanna find peaks!!!!!!!!!!!!!!
maxid_n1y = signal.argrelmax(n1y.to_numpy(), order = 45)

axes.plot(n1x.to_numpy()[maxid_n1y], n1y.to_numpy()[maxid_n1y], "yo", label = "Peak")


#Labeling other properties
axes.set_title("Curve of the 'un-saturated' spectrum", fontsize = 24)
axes.set_xlabel("Wavelength[$nm$]", fontsize = 20)
axes.set_ylabel("Intensity[(Arbitrary)$W/m^2$]", fontsize = 20)
axes.legend(fontsize = 18)

Peaks = n1x.to_numpy()[maxid_n1y]
Intensity_Peaks = n1y.to_numpy()[maxid_n1y]
print(Peaks[10], Intensity_Peaks[10])
print(Peaks[11], Intensity_Peaks[11])
print(Peaks[12], Intensity_Peaks[12])
print(Peaks[13], Intensity_Peaks[13])
print(Peaks[14], Intensity_Peaks[14])
print(Peaks[15], Intensity_Peaks[15])
```
