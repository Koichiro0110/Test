```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import matplotlib.patches as pt
from IPython.display import HTML

# We need these to animate on the notebook
```

```python
x = np.linspace(0, 4*np.pi, 1000)

fig1 = plt.figure(figsize = (9, 4))
ax = fig1.add_subplot(1, 1, 1)
for t in np.linspace(0, 20, 5):
    y = np.sin(x - 0.1*t)
    plt.plot(x, y)
```

```python
x = np.linspace(0, 4*np.pi, 1000)
omega = 1
num_frames = 100
speed = 20 # Speed of the animations in frames/sec.

fig2 = plt.figure(figsize = (9, 4))

#
# Setup the plotting limits.
plt.xlim(0, 4*np.pi)
plt.ylim(-1.1, 1.1)

line, = plt.plot([], [], color = "red") # Empty plot line in red.
# Calling this after the plot was created, causes it NOT to be plotted.
# We don't need it plotted, since we just want to show the video.

plt.close()

# Initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return(line,)

# This is the animation routine. It paints the frame for time t.
def animate(t):
    y = np.sin(x - omega*t)
    line.set_data(x, y)
    return(line,)

# We set a time array that is just long enough to have one complete wave cycle go by.
# We break it up into num_frames steps.
time = np.linspace(0, 2*np.pi/omega, num_frames)

# The next line computes the animation!
wave_ani = animation.FuncAnimation(fig2, animate, time, blit = True, interval = 1000/speed, init_func = init)

HTML(wave_ani.to_html5_video())
```

```python
from matplotlib import rc
rc('animation', html='html5')

x = np.linspace(0.,4*np.pi,1000)
omega = 1.
num_frames=100
speed=20.   # Speed of the animation in frames/sec. 
fig2 = plt.figure(figsize=(9,4))

#
# Setup the plotting limits.
plt.xlim(0,4*np.pi)
plt.ylim(-1.1,1.1)
# Here we setup the axes of the plot. We can set limits 
line, = plt.plot([],[],color="red")  # This is the empty plot line, in red.
dot1, =  plt.plot([],[],'o',color="green",markersize=10)
dot2, =  plt.plot([],[],'o',color="blue",markersize=15)

# Calling this after the plot was created, causes it NOT to be plotted.
# We don't need it plotted, since we just want to show the video
plt.close()
# initialization function: plot the background of each frame
def init():
    line.set_data([],[])
    dot1.set_data([],[])
    dot2.set_data([],[])
    return(line,dot1,dot2,)

# This is the animation routine. It paints the frame for time t.
def animate(t):
    y = np.sin(x - omega*t)
    ydot= np.sin(1 - omega*t)
    line.set_data(x,y)
    dot1.set_data([1],ydot)
    dot2.set_data([0.5*np.pi+omega*t],[1.])
    return(line,dot1,dot2,)

# We set a time array that is just long enough to have one complete wave cycle
# go by. We break it up into num_frame steps.
time = np.linspace(0.,2*np.pi/omega,num_frames) 
#
# The next line computes the animation.
wave_ani = animation.FuncAnimation(fig2,animate,time,blit=True,interval=1000/speed,init_func=init)
# This line, takes the animation and outputs it. Since we asked for HTML5, we get an HTML5 video in our
# browser screen. Choose either the jshtml, which uses java script to display the animation, or
# use html5, which creates an html5 video file. 
HTML(wave_ani.to_jshtml())
# HTML(wave_ani.to_html5_video())

```
