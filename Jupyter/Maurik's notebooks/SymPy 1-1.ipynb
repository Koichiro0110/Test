```python
from sympy import *
init_printing()
```

```python
x = Symbol("x") #SymbolのSは大文字！
```

```python
a = x**2 - 1
a
```

```python
y, z = symbols("y, z") #symbolsのsは小文字！
```

```python
a.subs(x, y+1)#x に y+1を代入
```

```python
a = (x + y - z)**6 #このままでは展開しない形のまま
a
```

```python
a = expand(a)  # expand()で展開させる
a
```

```python
degree(a, x) # degree(a, x)　→ a(x)のxについての次数
```

```python
collect(a, x) # collect(a, x) → aをxについて次数別で括る!
```

```python
a = factor(a) #factor(a)→aを因数分解する!
a
```

```python
a = (x**3 - y**3)/(x**2 - y**2) #このままでは約分してくれない
a
```

```python
cancel(a)  #cancel(a)→aを約分!
```

```python
a = y/(x - y) + x/(x + y)  #このままでは通分してくれない
a
```

```python
together(a) #together(a)→aを通分！
```

```python
simplify(a) #曖昧だけど何かシンプルにしてくれるらしい
```

```python
apart(a, x) # apart(a, x)→Partial fraction decomposition with respect to x, xについての部分分数分解
```

```python
b = a.subs({x:1, y:2}) # a.subs({x:k})→aにx=kを代入
c = a.subs({x:2})
b, c
```

```python
b.n() #b.n()→how much is it numerically? bの値は数値的にどの程度かを表示!
```

```python
sin(-x) #初等関数についてはSymPyが勝手に簡単にしてくれる！
```

```python
cos(pi/4), tan(5*pi/6)
```

```python
pi.n(100) #a.()の()の中身で表示させたい桁数を指定!
```

```python
log(1), log(E) # Eは自然対数の基準値を渡す
```

```python
exp(log(x)), log(exp(x))
#右側はxにならない,これはlog関数が複素数まで定義されているから!
#下記を参照
```

```python
n = log(exp(x))
n = n.subs(x, 2*pi*I) # Iは虚数単位!!
n
```

```python
sqrt(x)**2, sqrt(x**2)
#これも指数関数が複素数まで定義されているから
```

```python
f = sqrt(x**2)
f = f.subs(x, I)
f
```

```python
p, q = symbols("p, q", positive = True) #　positive = True → これでp, qを正の実数に限定！！
sqrt(p**2)
```

```python
n = Symbol("n", integer = True)
exp(2*pi*I*n)
```

```python
s = sin(pi*n/2)
k = s.rewrite(exp) #rewrite(f)でfの関数系を用いて表現させられる!!
s, k
```

```python
cos(x).rewrite(exp), exp(I*x).rewrite(cos)
```

```python
asin(x).rewrite(log) # asin は arcsin 又は　sin^(-1)を表す！
```

```python
trigsimp(2*sin(x)**2 + 3*cos(x)**2) #trigsimpは三角関数(trigonometric function)をシンプルにしてくれる
```

```python
expand_trig(sin(x-y)), expand_trig(sin(2*x)) #expand_trigは三角関数の和積の分解
```
