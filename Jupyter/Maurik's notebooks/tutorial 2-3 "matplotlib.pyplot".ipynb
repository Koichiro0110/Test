```python
def f(x):
    return x ** 2

#np.linspace(low, high, split width)
x_values = np.linspace(0, 1, 100)
y_values = f(x_values)

plt.plot(x_values, y_values)

plt.show()
```

```python
def f(x):
    return x ** 2

x_values = np.linspace(0, 1, 100)
y_values = f(x_values)

#plt.plot(x values, y values, color = 'colorname', linestyle = 'linestyle')
plt.plot(x_values, y_values, color = 'green', linestyle = '--')

plt.show()
```

```python
def f(x):
    return x ** 2

def g(x):
    return x ** 3 

x_values = np.linspace(0, 1, 100)
f_values = f(x_values)
g_values = g(x_values)

#first graph
plt.plot(x_values, f_values, color = 'green', linestyle = '--')
#second graph
plt.plot(x_values, g_values, color = 'red', linestyle = '-.')

plt.show()
```

```python
def f(a,x):
    return a * x ** 2

x_values = np.linspace(0, 1, 100)

for a in range(1, 20):
    y_values = f(a, x_values)
    plt.plot(x_values, y_values, color = 'red')

plt.show()
```

```python
def f(a,x):
    return a * x ** 2

x_values = np.linspace(0, 1, 100)

for a in range (1, 20):
    y_values = f(a, x_values)
    plt.plot(x_values, y_values, color = (a/20, 0.0, 0.0))

plt.show()
```

```python
def f(x):
    return x ** 2

x_values = np.linspace(0, 1, 100)
y_values = f(x_values)

plt.plot(x_values, y_values, color = (0.5, 0.0, 0.5))

plt.xlabel('donuts')
plt.ylabel('productivity')
plt.title('productivity v. donuts')

plt.show()

```

```python
x_values = np.linspace(0, 5 * np.pi, 200)
y_values = np.sin(x_values)

plt.plot(x_values, y_values, "green")

plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin(x) plotting')
plt.show()
```

```python
x_values = np.linspace(-3 * np.pi, 5 * np.pi, 200)
sine_values = np.sin(x_values)
cosine_values = np.cos(x_values)

plt.plot(x_values, sine_values, color = 'red', linestyle = '-')
plt.plot(x_values, cosine_values, color = 'green', linestyle = '--')

plt.xlabel('angle')
plt.ylabel('function value')
plt.show()
```

```python
def f(n, x):
    return np.sin(n * np.pi * x)

x_values = np.linspace(0, 1, 100)

for n in range(9):
    y_values = f(n+1, x_values)
    plt.plot(x_values, y_values, color = ((9-(n))/9, 0, n/9))
```
