```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = range(5) #create a list [0, 1, 2, 3, 4]
y = [q * q for q in x] #use list comprehension
plot = plt.plot(x, y) #name the plot
```

```python
x = np.linspace(0, 5, 100)
y = x * x
plot = plt.plot(x, y)
```

```python
# This creates a 97 figure space
fig = plt.figure(figsize = (9,7))

# the amount of white space to leave for the border
# [x_low, y_low, x_width, y_width]
# [0, 0, 1, 1] no space

#
axes = fig.add_axes([0, 0, 1, 1])

#
axes.plot(x, y, label = "$x^2$") #legend

axes.set_title('Plots of Polynomials', fontsize = 24)
axes.set_xlabel('x', fontsize = 16)
axes.set_ylabel('y', fontsize = 16)
axes.legend(fontsize = 18) #legend
```

```python
x = np.linspace(0, 4 * np.pi, 1000)
y = np.sin(x)

fig = plt.figure(figsize = (9,7))

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)

#x(ticks)
ax.set_xticks([i * np.pi/2 for i in range(9)])

#x""
ax.set_xticklabels([f"{i/2}$~\pi$" for i in range(9)], fontsize = 14)

#
ax.set_title("Sine Curve", fontsize = 24)

ax.set_xlabel('x', fontsize = 16)
ax.set_ylabel('y', fontsize = 16)

#
plt.yticks(fontsize = 14)
plt.show()
```

```python
x = np.linspace(-1, 1, 100)

y1 = x
y2 = x * x
y3 = x * x * x
y4 = x * x * x * x

fig = plt.figure(figsize = (10,8))

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y1, label = "$x$", color = "red")
axes.plot(x, y2, label = "$x^2$", color = "blue")
axes.plot(x, y3, label = "$x^3$", color = "green")
axes.plot(x, y4, label = "$x^4$", color = "turquoise")
# Note the use of $ formula $ - this typesets using LatTex.

axes.set_title('Plots of Polynomials', fontsize = 24)
axes.set_xlabel('x', fontsize = 16)
axes.set_ylabel('y', fontsize = 16)
axes.legend(fontsize = 18)
```

```python
x = np.linspace(0.1, 10, 1000)

y1 = x
y2 = x * x
y3 = x * x * x
y4 = x * x * x * x

fig = plt.figure(figsize = (10,8))

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y1, label = "$x$", color = "red")
axes.plot(x, y2, label = "$x^2$", color = "blue")
axes.plot(x, y3, label = "$x^3$", color = "green")
axes.plot(x, y4, label = "$x^4$", color = "turquoise")
# Note the use of $ formula $ - this typesets using LatTex.

axes.set_title('Log-Log plots of Polynomials', fontsize = 24)
axes.set_xlabel('x', fontsize = 16)
axes.set_ylabel('y', fontsize = 16)
axes.legend(fontsize = 18)

axes.semilogy()
axes.semilogx()
```
