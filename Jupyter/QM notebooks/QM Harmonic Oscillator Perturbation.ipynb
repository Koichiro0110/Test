```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import sympy  as sym
```

```python
ax = plt.figure().add_subplot(projection = '3d')
X, Y, Z = axes3d.get_test_data(0.05) # 適当なデータをモジュールからとってきている

# Plot the 3D surface
ax.plot_surface(X, Y, Z, edgecolor = 'royalblue', lw = 0.5, rstride = 8,cstride = 8, alpha = 0.3)
#alpha で透明にしている。
#edgecolorは明らかに色

ax.set(xlim = (-40,40), ylim = (-40, 40), zlim = (-100, 100),
      xlabel = 'X', ylabel = 'Y', zlabel = 'Z', title = 'aaa')

plt.show()
```

```python
ax = plt.figure().add_subplot(111, projection = '3d')

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
x, y = np.meshgrid(x, y)

f = 100
m = 1
k1 = 0.5
k2 = 1

V0 = 1/2*m*(f**2)*((x**2)+(y**2))

Vp1 = k1*m*(f**2)*x*y
V1 = V0 + Vp1

Vp2 = k2*m*(f**2)*x*y
V2 = V0 + Vp2

p0 = ax.plot_surface(x, y, V0, color = 'royalblue', edgecolor ='royalblue', lw = 0.5, rstride = 8, cstride = 8, alpha = 0.3)
p1 = ax.plot_surface(x, y, V1, color = 'mediumspringgreen', edgecolor = 'mediumspringgreen',lw = 0.5, rstride = 8, cstride = 8, alpha = 0.3)
p2 = ax.plot_surface(x, y, V2, color = 'khaki', edgecolor = 'khaki',lw = 0.5, rstride = 8, cstride = 8, alpha = 0.3)

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('V(x,y)')
ax.set_title('H.O. Perturbation')

p1.edgecolors2d = p1._edgecolor3d
p1.edgecolors2d = p1._facecolor3d



```
