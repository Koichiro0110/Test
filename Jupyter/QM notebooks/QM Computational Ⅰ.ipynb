```python
import numpy as np
from scipy import interpolate
import scipy.integrate as integrate
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#(1)
#(a)
#[i]

start = 0
stop = 1
N = 100

#Create the discretized x array (0≦x≦1)
x = np.linspace(start, stop, N)

#Create the discretized wave function φ_3 (0≦x≦1)
φ_3 = np.sqrt(2)*np.sin(3*np.pi*x)

#print(φ_3)
```

```python
#[ii]

#Define |φ_3|^2
SA_φ_3 = (np.abs(φ_3))**2

#Check φ_3 normalized
Normalization_φ_3 = integrate.trapz(SA_φ_3[0 : N], x[0 : N])

print("Σ|φ_3|^2 = "+str(Normalization_φ_3))
```

```python
#[iii]

#Define x*|φ_3|^2
el_expectation_value_x = x*SA_φ_3

#Calculate <x> = Σ x*|φ_3|^2
expectation_value_x = integrate.trapz(el_expectation_value_x[0 : N], x[0 : N])

print("<x> = "+str(expectation_value_x))

#Define x^2*|φ_3|^2
      
el_expectation_value_xx = x*x*SA_φ_3

#Calculate <x^2> = Σ x^2*|φ_3|^2
expectation_value_xx = integrate.trapz(el_expectation_value_xx[0 : N], x[0 : N])

print("<x^2> = "+str(expectation_value_xx))

#Calculate σ_x = √(<x^2> - <x>^2)
σ_x = np.sqrt(expectation_value_xx - (expectation_value_x)**2)

print("σ_x = "+str(σ_x))
```

```python
#[iv]

#Define d(φ_3)/dx
diff_φ_3 = np.zeros_like(φ_3)
diff_φ_3[0 : -1] = np.diff(φ_3)/np.diff(x)
diff_φ_3[-1] = (φ_3[-1] - φ_3[-2])/(x[-1] - x[-2])

#Define -i*(h/2π)*(φ_3)*d(φ_3)/dx
hbar = 1.0547*(10**(-34))
el_expectation_value_p = (0 + 1j)*hbar*φ_3*diff_φ_3

#Take sum
expectation_value_p = integrate.trapz(el_expectation_value_p[0 : N], x[0 : N])

print("<p> = "+str(expectation_value_p))

#Define dd(φ_3)/ddx
ddiff_φ_3 = np.zeros_like(diff_φ_3)
ddiff_φ_3[0 : -1] = np.diff(diff_φ_3)/np.diff(x)
ddiff_φ_3[-1] = (diff_φ_3[-1] - diff_φ_3[-2])/(x[-1] - x[-2])

#Define -i*(h/2π)*(φ_3)*dd(φ_3)/ddx
el_expectation_value_pp = -1*(hbar**2)*φ_3*ddiff_φ_3

#Take sum
expectation_value_pp = integrate.trapz(el_expectation_value_pp[0 : N], x[0 : N])

print("<p^2> = "+str(expectation_value_pp))

σ_p = np.sqrt(expectation_value_pp - (expectation_value_p)**2)

print("σ_p = "+str(σ_p))
```

```python
#** [iv] by using spline evaluation
spline_rep = interpolate.splrep(x, φ_3)
dφ_3_s = interpolate.splev(x, spline_rep, der = 1)

el_expectation_value_p_s = -(0 + 1j)*hbar*φ_3*dφ_3_s

expectation_value_p_s = integrate.trapz(el_expectation_value_p_s[0 : N], x[0 : N])

print("<p> = "+str(expectation_value_p_s))

ddφ_3_s = interpolate.splev(x, spline_rep, der = 2)

el_expectation_value_pp_s = -(hbar**2)*φ_3*ddφ_3_s

expectation_value_pp_s = integrate.trapz(el_expectation_value_pp_s[0 : N], x[0 : N])

print("<p^2> = "+str(expectation_value_pp_s))

σ_p_s = np.sqrt(expectation_value_pp_s - (expectation_value_p_s)**2)

print("σ_p = "+str(σ_p_s))
```

```python
#(2)
#(a)

#Define c_n as a function of n
def c(n):
    return 4*np.sqrt(6)/((n*np.pi)**2)*((-1)**((n-1)/2))*1/2*(1 - ((-1)**n))

print("c_1 = "+str(c(1)), "c_2 = "+str(c(2)), "c_3 = "+str(c(3)))
```

```python
#(b)

# c_n is a decaying function with n, so I take the sum by enough large number 10000 to justify the unity.
x = range(1, 10000, 1)
sum_c_n = 0
for k in x:
    sum_c_n = sum_c_n + (np.abs(c(k)**2))
    
print("Σc_n = "+str(sum_c_n))
```

```python
#(c)

#I set a as a=1
def E(n):
    return ((n*np.pi)**2)/2

print("E_1 = "+str(E(1)), "E_2 = "+str(E(2)), "E_3 = "+str(E(3)))
```

```python
#(d)
#0<x<1
def ψ(x, t):
    y = range(1, 10000, 1)
    sum_ψ_n = 0
    for k in y:
        sum_ψ_n = sum_ψ_n + c(k)*np.sqrt(2)*np.sin(k*np.pi*x)*np.exp(-(0+1j)*E(k*t))
        
    return sum_ψ_n

print(ψ(1, 1))
```

```python
#(e)
#(i)

#0<x<1(inside the well)

#0<t<1

#Create a figure
fig = plt.figure(figsize = (10, 8))
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.set_title(r"Re($\psi(x,t)$)", fontsize = 30)
ax1.set_xlabel(r"$x$", fontsize = 20)
ax1.set_ylabel(r"Re($\psi(x,t)$)", fontsize = 20)

cmap = plt.get_cmap("Blues")

#Plot
el_position = np.linspace(0, 1, 100)
timepoints = np.linspace(0, 1, 10)
y = np.zeros_like(el_position)
for k in range(len(timepoints)):
    grad = (len(timepoints) - k)/len(timepoints)
    for i in range(len(el_position)):
        y[i] = np.real(ψ(el_position[i], timepoints[k]))
    ax1.plot(el_position, y, label = "0<t<1, t = "+str(k), color = cmap(grad)) #Make a gradation for time evolution

ax1.legend()
```
