```python
import numpy as np
from scipy import special
import matplotlib.pyplot as plt
```

```python
# Define Legendre Functions
# Take only odd integers
def P(n,x):
    a = special.eval_legendre(2*n+1,x)
    return a
def C(n,x):
    a = (-1)**(n)*(4*n+3)*special.factorial2(2*n+1)/((2*n+1)*(2*n+2)*special.factorial2(2*n))
    return a
    
#要修正

def LegendreExpansion(m,x):
    a = 0
    for n in range(m):
        a += C(n,x) * P(n,x)    
    return a
        
    
```

```python
x = np.linspace(-1, 1, 1000)

y5 = LegendreExpansion(5,x)
y20 = LegendreExpansion(20,x)
y100 = LegendreExpansion(100,x)
y150 = LegendreExpansion(150,x)

fig = plt.figure(figsize = (10,8))
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y5, label = "N = 5")
axes.plot(x, y20, label = "N = 20")
axes.plot(x, y100, label = "N = 100")
axes.plot(x, y150, label = "N = 150")
axes.set_title('Plots of Legendre Expansion by The Nth Term', fontsize = 22)
axes.set_xlabel('x', fontsize = 16)
axes.set_ylabel('$\Sigma~C_N~P_{2N+1}(x)$ ', fontsize = 16)
axes.legend(fontsize = 18)
```
